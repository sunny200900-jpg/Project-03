## Bivariate Analysis


Example 1: Living Area vs Sale Price (Numeric–Numeric)

Variables:

GrLivArea → Living area (sq. ft)

SalePrice → House price

Purpose

To understand how house size affects price.


```python
plt.scatter(df['GrLivArea'], df['SalePrice'])
plt.xlabel("Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.title("Bivariate Analysis: Living Area vs Sale Price")
plt.show()
```

##Observation

Strong positive relationship

As living area increases, sale price increases

Conclusion

Living area is a key determinant of house price

Insight:

Linear or non-linear relationship

Correlation direction

Outliers

##Example 2: Number of Bedrooms vs Sale Price (Categorical–Numeric)

Variables:

BedroomAbvGr → Number of bedrooms

SalePrice → House price

Purpose

To analyze how bedroom count influences pricing.

```python
df.groupby('BedroomAbvGr')['SalePrice'].mean().plot(kind='bar')
plt.xlabel("Number of Bedrooms")
plt.ylabel("Average Sale Price")
plt.title("Bivariate Analysis: Bedrooms vs Sale Price")
plt.show()
```

##Observation

Prices increase with bedrooms up to a point

After that, the increase is marginal

Conclusion

More bedrooms increase value, but with diminishing returns

##Example 3: Garage Cars vs Sale Price (Categorical–Numeric)

Variables:

GarageCars → Garage capacity

SalePrice → House price

```python
df.groupby('GarageCars')['SalePrice'].mean().plot(kind='bar')
plt.xlabel("Garage Capacity")
plt.ylabel("Average Sale Price")
plt.title("Bivariate Analysis: Garage Capacity vs Sale Price")
plt.show()
```

##Observation

Houses with 2–3 car garages are priced higher

Conclusion

Garage space significantly impacts house valuation

Insight:

Category comparison

Mean trends

##xample 4: Overall Quality vs Sale Price (Ordinal–Numeric)

Variables:

OverallQual → Quality rating (1–10)

SalePrice → House price

```python
plt.scatter(df['OverallQual'], df['SalePrice'])
plt.xlabel("Overall Quality")
plt.ylabel("Sale Price")
plt.title("Bivariate Analysis: Quality vs Sale Price")
plt.show()
```

##Observation

Strong upward trend

Higher quality = higher price

Conclusion

Overall quality has a strong positive effect on pricing

## Multivariate analysis

##Example 1: Living Area + Bedrooms + Bathrooms vs Sale Price

Variables involved (4 variables):

GrLivArea (size)

BedroomAbvGr (bedrooms)

FullBath (bathrooms)

SalePrice (target)

🔹 Purpose

To understand how multiple house features together influence price.

🔹 Code (Scatter with Color Encoding)

```python
sns.scatterplot(
    x='GrLivArea',
    y='SalePrice',
    hue='BedroomAbvGr',
    size='FullBath',
    data=df
)

plt.title("Multivariate Analysis: Area, Bedrooms, Bathrooms vs Price")
plt.show()
```

##Observation

Larger houses with more bedrooms and bathrooms are priced higher

Color and size show combined influence

🔹 Conclusion

House price depends on multiple features simultaneously

##Example 2: Correlation Heatmap (Multiple Numeric Variables)

Variables:

SalePrice

GrLivArea

OverallQual

GarageCars

TotalBsmtSF

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[['SalePrice','GrLivArea','OverallQual','GarageCars','TotalBsmtSF']].corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Multivariate Correlation Analysis")
plt.show()
```

##Observation

Strong correlation between SalePrice and:

OverallQual

GrLivArea

GarageCars

🔹 Conclusion

Multiple variables jointly explain price variation


Insight:

Strength & direction of relationships

Feature selection

##Example 3: Boxplot with Multiple Categories

Variables (3 variables):

SalePrice (numeric)

Neighborhood (categorical)

OverallQual (categorical)

```python
plt.figure(figsize=(10,5))
sns.boxplot(
    x='Neighborhood',
    y='SalePrice',
    hue='OverallQual',
    data=df
)
plt.xticks(rotation=90)
plt.title("Multivariate Analysis: Location & Quality vs Price")
plt.show()
```

##Observation

Prices vary significantly by location and quality

High-quality houses command premium prices across neighborhoods
